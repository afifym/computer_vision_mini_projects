```python
import cv2
import numpy as np
import helper_functions as hf
```

```python
fire_orig = cv2.imread('fire.jpg')
cold_orig = cv2.imread('cold.jpg')
cold_orig = hf.crop_center(cold_orig, 100)
```

```python
shape = (2048, 2048)
pyr_depth = 9

fire = cv2.resize(fire_orig, shape)
cold = cv2.resize(cold_orig, shape)

def gaussian_pyramid(image, depth):
    downres = image.copy()
    g_pyramid = [downres]
    for i in range(depth):
        downres = cv2.pyrDown(downres)
        g_pyramid.append(downres)

    return g_pyramid

def laplacian_pyramid(g_pyramid, depth):
    lp_pyramid = [g_pyramid[-1]]
    for i in range(depth, 0, -1):
        upres = cv2.pyrUp(g_pyramid[i])
        lp = cv2.subtract(g_pyramid[i-1], upres)
        lp_pyramid.append(lp)
    
    return lp_pyramid
```

```python
g_fire = gaussian_pyramid(fire, pyr_depth)
g_cold = gaussian_pyramid(cold, pyr_depth)

lp_fire = laplacian_pyramid(g_fire, pyr_depth)
lp_cold = laplacian_pyramid(g_cold, pyr_depth)

fire_cold_pyramid = []
for fire_lap, cold_lap in zip(lp_fire, lp_cold):
    width, height, _ = fire_lap.shape
    laplac = np.hstack((fire_lap[:, :width//2], cold_lap[:, height//2:]))
    fire_cold_pyramid.append(laplac)
    

recons = fire_cold_pyramid[0]
for i in range(1, pyr_depth):
    recons = cv2.pyrUp(recons)
    recons = cv2.add(fire_cold_pyramid[i], recons)
    
hf.plt_imshow_bgr(recons)
```
